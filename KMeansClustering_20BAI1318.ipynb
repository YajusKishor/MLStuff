```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt

iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
sns.heatmap(df.corr(), annot=True, linecolor='white', linewidths=2)
```

```python
df.hist()
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3).fit(df)
```

```python
kmeans.labels_
```

```python
# Visualize the clusters with respect to the first two columns
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=kmeans.labels_)
plt.show()
```

```python
# Compute the accuracy of the clustering
from sklearn.metrics import accuracy_score

accuracy_score(kmeans.labels_, iris.target)
print(kmeans.labels_, iris.target)
```
