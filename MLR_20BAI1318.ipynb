```python
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn import linear_model, metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("FuelConsumption.csv")
print(df.shape)
print(df.info())
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
regress = linear_model.LinearRegression()
df2 = cdf.values
train_x =   np.asanyarray(cdf[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
train_y = np.asanyarray(cdf[['CO2EMISSIONS']])

print(train_x.shape)
print(train_y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

ms = MinMaxScaler()
train_x = ms.fit_transform(train_x)
```

```python
regress.fit(train_x, train_y)

print('Coefficients: ', regress.coef_)
print('Intercept: ', regress.intercept_)
```

```python
y_pred = regress.predict(train_x)
for i in range(0, len(train_x)):
    print(train_y[i], y_pred[i])
```

```python
print('Mean Absolute Error: ', metrics.mean_absolute_error(train_y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(train_y, y_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(train_y, y_pred)))
```
